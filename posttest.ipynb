```python
# Mengimpor pustaka matplotlib untuk membuat grafik
import matplotlib.pyplot as plt

# Membuat daftar koordinat sumbu x
x = [1, 2, 3, 4, 5]

# Membuat daftar koordinat sumbu y
y = [2, 4, 6, 8, 10]

# Membuat plot titik (scatter plot) dari koordinat x dan y
plt.scatter(x, y, color='red')

# Memberikan judul pada grafik
plt.title("Titik Piksel")

# Memberikan label pada sumbu X
plt.xlabel("Sumbu X")

# Memberikan label pada sumbu Y
plt.ylabel("Sumbu Y")

# Menampilkan hasil grafik di layar
plt.show()
```
